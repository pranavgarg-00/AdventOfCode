```python
%reset -f

def numcombs(patt, nums):
    c = patt.count("#") 
    d = patt.count("?")
    s = sum(nums)
    if c + d < s or c > s:
        return 0
    if c + d == 0:
        return 1
    
    if patt[0] == '.':   
        return numcombs(patt[1:], nums)
    
    if patt[0] == '#':
        a = patt[:nums[0]]
        if a.count('#') + a.count('?') != nums[0]:
            return 0
        if nums[0] < len(patt) and patt[nums[0]] == '#':
            return 0
        return numcombs(patt[nums[0] + 1:], nums[1:])
    
    return numcombs(["#"] + patt[1:], nums) + numcombs(patt[1:], nums) 
        
    

with open("day12.txt") as f:   
    s = f.read()
    L = s.split("\n")
    answers = []
    for i in range(len(L)):
        left, right = L[i].split()
        nums = list(map(int, right.split(',')))
        patt = list(left)
        answers.append(numcombs(patt, nums))
    print(sum(answers))
```

```python
%reset -f

smudge = {'.':'#', '#':'.'}

def getreflection(mat):
    answers = set()
    
    # horizontal mirror
    for i in range(1, len(mat)):
        minsize = min(i, len(mat) - i)
        a = mat[max(0, i-minsize):i][::-1]
        b = mat[i:min(len(mat), i+minsize)]
        if a == b:
            answers.add((i,0))
    # vertical mirror

    tempmat = [['.']*len(mat) for x in range(len(mat[0]))]
    for i in range(len(mat)):
        for j in range(len(mat[0])):
            tempmat[j][i] = mat[i][j]

    for i in range(1, len(tempmat)):
        minsize = min(i, len(tempmat) - i)
        a = tempmat[max(0, i-minsize):i][::-1]
        b = tempmat[i:min(len(tempmat), i+minsize)]
        if a == b:
            answers.add((0,i))
            
    return answers
    

with open("day13.txt") as f:   
    s = f.read()
    L = s.split("\n\n")
    finalanswers = 0
    for T in range(len(L)):
        temp = list(map(list, L[T].split()))
        mainref = getreflection(temp)
        flag = False
        for i in range(len(temp)):
            if flag:
                break
            for j in range(len(temp[0])):
                temp[i][j] = smudge[temp[i][j]]
                ans = getreflection(temp)
                if ans.difference(mainref):
                    x = list(ans.difference(mainref))[0]
                    flag = True
                    finalanswers += (100*x[0] + x[1])
                    break
                temp[i][j] = smudge[temp[i][j]]
            
    print(finalanswers)
```

```python
%reset -f
def runcycle(mat):
    #north
    for j in range(len(mat[0])):
        for i in range(len(mat)):
            if mat[i][j] == 'O':
                tempi = i
                while tempi-1 >= 0 and mat[tempi-1][j] == '.':
                    tempi -= 1
                mat[tempi][j], mat[i][j] = mat[i][j], mat[tempi][j]
    
    #west
    for i in range(len(mat)):
        for j in range(len(mat[0])):
            if mat[i][j] == 'O':
                tempj = j
                while tempj-1 >= 0 and mat[i][tempj-1] == '.':
                    tempj -= 1
                mat[i][tempj], mat[i][j] = mat[i][j], mat[i][tempj]
    
    #south
    for j in range(len(mat[0])):
        for i in reversed(range(len(mat))):
            if mat[i][j] == 'O':
                tempi = i
                while tempi+1 < len(mat) and mat[tempi+1][j] == '.':
                    tempi += 1
                mat[tempi][j], mat[i][j] = mat[i][j], mat[tempi][j]
    
    #east
    for i in range(len(mat)):
        for j in reversed(range(len(mat[0]))):
            if mat[i][j] == 'O':
                tempj = j
                while tempj+1 < len(mat[0]) and mat[i][tempj+1] == '.':
                    tempj += 1
                mat[i][tempj], mat[i][j] = mat[i][j], mat[i][tempj]
    
    return mat

with open("day14.txt") as f:   
    s = f.read()
    L = s.split("\n")
    mat = list(map(list, L))
    visited = set()
    visited.add(tuple(map(tuple, mat)))
    visitedlist = []
    visitedlist.append(tuple(map(tuple, mat)))
    ctr = 0
    frozenmat = None
    while True:
        ctr += 1
        mat = runcycle(mat)
        frozenmat = tuple(map(tuple, mat))
        if frozenmat in visited:
            print(ctr)
            break
        else:
            visited.add(frozenmat)
            visitedlist.append(frozenmat)
    print(visitedlist.index(frozenmat))
    
    moreapplies = (1000000000 - ctr)%(ctr - visitedlist.index(frozenmat))
    for i in range(moreapplies):
        mat = runcycle(mat)
    ans = 0
    for i in range(len(mat)):
        ans += (mat[i].count('O'))*(len(mat) - i)
    print(ans)
```

```python
%reset -f
def HASH(inp):
    cur = 0
    for i in range(len(inp)):
        cur = (17*(cur + ord(inp[i])))%256
    return cur

with open("day15.txt") as f:   
    s = f.read()
    s = s.replace("\n", "")
    L = s.split(",")
    lenses = [{} for i in range(256)]
    ans = 0
    for i in range(len(L)):
        if '-' in L[i]:
            key = L[i][:-1]
            keyhash = HASH(key)
            lenses[keyhash].pop(key, None)
        else:
            key = L[i][:-2]
            keyhash = HASH(key)
            lenses[keyhash][key] = int(L[i][-1])
    
    for i in range(len(lenses)):
        for j, key in enumerate(lenses[i]):
            ans += (i+1)*(j+1)*lenses[i][key]
        
    print(ans)
```
