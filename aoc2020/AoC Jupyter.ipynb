```python
def getadjacents(i, j, isize, jsize, grid):
    adjacents = []
    k = 1
    while(i-k>=0 and j-k>=0):
        if not grid[i-k][j-k] == '.':
            adjacents.append([i-k, j-k])
            break
        k += 1
    k = 1
    while(i-k>=0):
        if not grid[i-k][j] == '.':
            adjacents.append([i-k, j])
            break
        k += 1
    k = 1
    while(i-k>=0 and j+k<jsize):
        if not grid[i-k][j+k] == '.':
            adjacents.append([i-k, j+k])
            break
        k += 1
    k = 1
    while(j-k>=0):
        if not grid[i][j-k] == '.':
            adjacents.append([i, j-k])
            break
        k += 1
    k = 1
    while(j+k<jsize):
        if not grid[i][j+k] == '.':
            adjacents.append([i, j+k])
            break
        k += 1
    k = 1
    while(i+k<isize and j-k>=0):
        if not grid[i+k][j-k] == '.':
            adjacents.append([i+k, j-k])
            break
        k += 1
    k = 1
    while(i+k<isize):
        if not grid[i+k][j] == '.':
            adjacents.append([i+k, j])
            break
        k += 1
    k = 1
    while(i+k<isize and j+k<jsize):
        if not grid[i+k][j+k] == '.':
            adjacents.append([i+k, j+k])
            break
        k += 1
    return adjacents
    
```

```python
def maskupdate(memmask, number):
    bits = list(bin(number)[2:])
    while len(bits) < 36:
        bits.insert(0, '0')
    for i in range(len(memmask)):
        if memmask[i] == 'X':
            continue
        bits[i] = memmask[i]
    return int("".join(bits), 2)

def masklist(memmask, address):
    finallist = []
    list1 = list(bin(address)[2:])
    while len(list1) < 36:
        list1.insert(0, '0')
    for i in range(len(memmask)):
        if memmask[i] == '0':
            continue
        list1[i] = memmask[i]
    flag = 0
    finallist.append(list1)
    while(flag == 0):
        flag = 1
        temp = []
        for i in range(len(finallist)):
            for j in range(len(finallist[i])):
                if finallist[i][j] == 'X':
                    flag = 0
                    tlist1 = finallist[i].copy()
                    tlist2 = finallist[i].copy()
                    tlist1[j] = '0'
                    tlist2[j] = '1'
                    temp.append(tlist1)
                    temp.append(tlist2)
                    break
        if len(temp) > 0:
            finallist = temp
    finalnums = []
    for i in finallist:
        finalnums.append(int("".join(i), 2))
    return finalnums

print(masklist("000000000000000000000000000000X1001X", 42))
    
        
        
```

```python
def evaluate(exstring):
    if not '(' in exstring:
        x = list(exstring.split())
        for i in range(len(x)):
            if x[i].isnumeric():
                x[i] = int(x[i])
        value = x[0]
        for i in range(2, len(x), 2):
            if x[i-1] == '+':
                value += x[i]
            elif x[i-1] == '*':
                value *= x[i]
        return value
    a = exstring.index('(')
    b = a
    stack = 1
    while stack > 0:
        b+=1
        if exstring[b] == '(':
            stack += 1
        elif exstring[b] == ')':
            stack -= 1    
    return evaluate(exstring[:a] + " " + str(evaluate(exstring[a+1:b])) + " " + exstring[b+1:])

def evaluate2(exstring):
    if not '(' in exstring:
        x = list(exstring.split())
        while '+' in x:
            a = x.index('+')
            x[a] = str(int(x[a-1]) + int(x[a+1]))
            x.pop(a+1)
            x.pop(a-1)
        return eval("".join(x))
    a = exstring.index('(')
    b = a
    stack = 1
    while stack > 0:
        b+=1
        if exstring[b] == '(':
            stack += 1
        elif exstring[b] == ')':
            stack -= 1    
    return evaluate2(exstring[:a] + " " + str(evaluate2(exstring[a+1:b])) + " " + exstring[b+1:])

f = open('day18.txt')
s = f.read()
eqs = list(s.split("\n"))
evals = []
for i in range(len(eqs)):
    evals.append(evaluate2(eqs[i]))
print(sum(evals))
```

```python
f = open('day22.txt')
s = f.read()
p1, p2 = s.split('\n\n')
p1cards = list(p1.split('\n'))[1:]
p2cards = list(p2.split('\n'))[1:]
for i in range(len(p1cards)):
    p1cards[i] = int(p1cards[i])
    p2cards[i] = int(p2cards[i])

def game(p1cards, p2cards):
    rounds = set()
    while len(p1cards) > 0 and len(p2cards) > 0:
        if (tuple(p1cards), tuple(p2cards)) in rounds:
            print(p1cards)
            return 1
        else:
            rounds.add((tuple(p1cards), tuple(p2cards)))
        a = p1cards.pop(0)
        b = p2cards.pop(0)
        if len(p1cards) >= a and len(p2cards) >= b:
            x = game(p1cards.copy()[:a], p2cards.copy()[:b])
            if x == 1:
                p1cards.append(a)
                p1cards.append(b)
            else:
                p2cards.append(b)
                p2cards.append(a)
        else:
            if a > b:
                p1cards.append(a)
                p1cards.append(b)
            else:
                p2cards.append(b)
                p2cards.append(a)
    if len(p1cards) == 0:
        print(p2cards)
        return -1
    else:
        print(p1cards)
        return 1
print(game(p1cards, p2cards))
```

```python
def blackneighbors(tiles, loc):
    ctr = 0
    if tiles.get((loc[0]+1, loc[1]), 1) == -1:
        ctr += 1
    if tiles.get((loc[0]-1, loc[1]), 1) == -1:
        ctr += 1
    if tiles.get((loc[0]+0.5, loc[1]+1), 1) == -1:
        ctr += 1
    if tiles.get((loc[0]+0.5, loc[1]-1), 1) == -1:
        ctr += 1
    if tiles.get((loc[0]-0.5, loc[1]+1), 1) == -1:
        ctr += 1
    if tiles.get((loc[0]-0.5, loc[1]-1), 1) == -1:
        ctr += 1
    return ctr
def neighbors(loc):
    n = []
    n.append((loc[0]+1, loc[1]))
    n.append((loc[0]-1, loc[1]))
    n.append((loc[0]+0.5, loc[1]+1))
    n.append((loc[0]+0.5, loc[1]-1))
    n.append((loc[0]-0.5, loc[1]+1))
    n.append((loc[0]+0.5, loc[1]-1))
    return n
for i in range(100):
    temp = {}
    for i in tiles:
        if not i in temp:
            if tiles[i] == 1:
                if blackneighbors(tiles, i) == 2:
                    temp[i] = -1
                else:
                    temp[i] = 1
            else:
                a = blackneighbors(tiles, i)
                if a == 0 or a > 2:
                    temp[i] = 1
                else:
                    temp[i] = -1
        for j in neighbors(i):
            if not j in temp:
                if tiles.get(j, 1) == 1:
                    if blackneighbors(tiles, j) == 2:
                        temp[j] = -1
                    else:
                        temp[j] = 1
                else:
                    a = blackneighbors(tiles, j)
                    if a == 0 or a > 2:
                        temp[j] = 1
                    else:
                        temp[j] = -1
    ctr = 0
    for i in temp:
        if temp[i] == -1:
            ctr+=1
    print(ctr)
    tiles = temp
```
